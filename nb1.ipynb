# 5 A more difficult Classification Problem

```python
[Xtr2,Ytr2] =  generate_dataset_classification(1000,20,True)
Ytr2 = utils.to_categorical(Ytr2, num_classes=3)
```

```python
model3 = Sequential()
model3.add(Dense(3, activation="softmax"))
model3.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
model3.fit(Xtr2, Ytr2, epochs= 100, batch_size= 32, verbose= 0)
```

The linear classification overfits the data: it has around 75% accuracy on the training set and only 40% on the testing set. We feel the need to create a more efficient model.

```python
model4 = Sequential()
model4.add(Conv2D(filters=16,kernel_size=(5,5),activation = "relu"))
model4.add(MaxPool2D((2,2)))
model4.add(Conv2D(filters = 32, kernel_size=(5,5),activation="relu"))
model4.add(MaxPool2D((2,2)))
model4.add(Flatten())
model4.add(Dense(32,activation="relu"))
model4.add(Dense(3,activation="softmax"))

model4.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
model4.fit(Xtr2.reshape(Xtr2.shape[0],72,72,1), Ytr2, epochs= 15, batch_size= 32, verbose= 1)
```

```python
[Xtest2, Ytest2] = generate_dataset_classification(600,20,True)
Ytest2 = utils.to_categorical(Ytest2, num_classes=3)
e3 = model3.evaluate(Xtest2,Ytest2)[1]
e4 = model4.evaluate(Xtest2.reshape(Xtest2.shape[0],72,72,1),Ytest2)[1]

print(e3,e4)
```

We managed to reach 92% test-set accuracy with the convolutional architecture, which is a good improvement of the linear classifier.

# 6 A regression problem

### Generation of the functions that will split the data. The purpose is to split the corners to understand them better. We choose to split the furthest right corner, the furthest left conner and the middle corner.

```python
def gen1(Y):
    U = Y[:,np.array([0,2,4])]
    V = Y[:,np.array([1,3,5])]
    U_1= U[range(U.shape[0]),U.argmax(1)]
    V_1 = V[range(U.shape[0]),U.argmax(1)]
    return(np.concatenate((U_1[:,np.newaxis],V_1[:,np.newaxis]), axis = 1))

def gen2(Y):
    U = Y[:,np.array([0,2,4])]
    V = Y[:,np.array([1,3,5])]
    U_1= U[range(U.shape[0]),U.argmin(1)]
    V_1 = V[range(U.shape[0]),U.argmin(1)]
    return(np.concatenate((U_1[:,np.newaxis],V_1[:,np.newaxis]), axis = 1))

def gen3(Y):
    U = Y[:,np.array([0,2,4])]
    V = Y[:,np.array([1,3,5])]
    
    ml = (U == U)
    indexes = np.kron(np.arange(U.shape[1]),np.ones((U.shape[0],1)))
    for k in range(U.shape[0]):
        ml[k] =(indexes[k] != U.argmax(1)[k]) 
        ml[k] = ml[k] & (indexes[k] != U.argmin(1)[k])
    return np.concatenate((U[ml][:,np.newaxis],V[ml][:,np.newaxis]), axis = 1)
```

```python
pui = 6

inputs = Input(shape=(72,72,1))
conv1a = Conv2D(2**pui,3,padding= "same",activation = 'relu')(inputs)
conv1b = Conv2D(2**pui,3,padding= "same",activation = 'relu')(conv1a)
pool1 = MaxPool2D(pool_size=(2, 2))(conv1b)

conv2a = Conv2D(2**(pui+1), 3,padding= "same", activation = 'relu')(pool1)
conv2b = Conv2D(2**(pui+1), 3,padding= "same", activation = 'relu')(conv2a)
pool2 = MaxPool2D(pool_size=(2, 2))(conv2b)

conv3a = Conv2D(2**(pui+2), 3,padding= "same", activation = 'relu')(pool2)
conv3b = Conv2D(2**(pui+2), 3,padding= "same", activation = 'relu')(conv3a)
pool3 = MaxPool2D(pool_size=(2, 2))(conv3b)

up4 = UpSampling2D(size = (2,2))(pool3)
conv4a = Conv2D(2**(pui+1),2,padding= "same",activation="relu")(up4)
conv4b = Conv2D(2**(pui+1),3,padding= "same",activation='relu')(conv4a)
conv4c = Conv2D(2**(pui+1),3,padding= "same",activation="relu")(conv4b)

up5 = UpSampling2D(size = (2,2))(conv4c)
conv5a = Conv2D(2**pui,2,padding= "same",activation="relu")(up5)
merge5 = merge.concatenate([conv2b,conv5a], axis = 3)
conv5b = Conv2D(2**pui,3,padding= "same",activation='relu')(merge5)
conv5c = Conv2D(2**pui,3,padding= "same",activation="relu")(conv5b)

up6 = UpSampling2D(size = (2,2))(conv5c)
conv6a = Conv2D(2**(pui-1),3,padding = "same", activation = "relu")(up6)
merge6 = merge.concatenate([conv1b,conv6a], axis = 3)
conv6b = Conv2D(2**(pui-1),3,padding = "same", activation = "relu")(merge6)
conv6c = Conv2D(1,1,activation = "sigmoid")(conv6b)

t7 = time.time()
model7 = Model(inputs = inputs, outputs = conv6c)
model7.compile(optimizer = "adam", loss = "binary_crossentropy")
model7.fit(Xtr7.reshape(Xtr7.shape[0],72,72,1), Ytr7.reshape(Ytr7.shape[0],72,72,1), epochs=10,batch_size=32,verbose=1)
print(time.time()-t7)
```
